```python
#function to collect songs in album into array
def collectSongs(song_names, rapper):
    lyrics = []
    for song_name in song_names:
        try:
            song = genius.search_song(song_name, rapper)
            if song is not None:
                lyrics.append(song)
        except:
            print(song_name)
    return lyrics
```

```python
# raw data dictionary
raw_lyrics_array = {}

# fill raw_data dictionary with song lyrics
for rapper in rappers:
    if rapper == "Eminem":
        raw_lyrics_array[rapper] = collectSongs(eminem_song_names, rapper)
    elif rapper == 'Jay Z':
        raw_lyrics_array[rapper] = collectSongs(jayz_song_names, rapper)
    elif rapper == 'Kanye West':
        raw_lyrics_array[rapper] = collectSongs(kanye_song_names, rapper)
    elif rapper == 'Kendrick Lamar':
        raw_lyrics_array[rapper] = collectSongs(kendrick_song_names, rapper)
    elif rapper == 'Tupac':
        raw_lyrics_array[rapper] = collectSongs(tupac_song_names, rapper)
    else:
        print("No match")
```

```python
# function to collect songs in album into one text
def collectSongsToOneText(song_names, rapper):
    text = ""
    for song_name in song_names:
        song = genius.search_song(song_name, rapper)
        if song is not None:
            text += song.lyrics + "\n";
    return text
```

```python
# raw data dictionary
raw_data = {}

# fill raw_data dictionary with song lyrics
for rapper in rappers:
    if rapper == "Eminem":
        raw_data[rapper] = collectSongsToOneText(eminem_song_names, rapper)
    elif rapper == 'Jay Z':
        raw_data[rapper] = collectSongsToOneText(jayz_song_names, rapper)
    elif rapper == 'Kanye West':
        raw_data[rapper] = collectSongsToOneText(kanye_song_names, rapper)
    elif rapper == 'Kendrick Lamar':
        raw_data[rapper] = collectSongsToOneText(kendrick_song_names, rapper)
    elif rapper == 'Tupac':
        raw_data[rapper] = collectSongsToOneText(tupac_song_names, rapper)
    else:
        print("No match")
```
