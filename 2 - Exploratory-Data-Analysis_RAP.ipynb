```python
# Read in the document-term matrix
import pandas as pd

data = pd.read_pickle('//content/drive/MyDrive/project_rap/clean_data_dtm.pkl')          
data = data.transpose()
data
```

```python
# Find the top 30 words said by each rap artist
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(30)
    top_dict[c]= list(zip(top.index, top.values))

top_dict
```

```python
top_dict_df = pd.DataFrame.from_records(top_dict)
#top_dict_df.columns = ['words']
top_dict_df = top_dict_df.sort_index()

top_dict_df

# pickle collected dataset for later use
#raw_data_df.to_pickle("raw_data_df.pkl")
```

```python
import matplotlib.pyplot as plt
```

```python
for k in top_dict.keys():
    keys = []
    values = []
    i = 0
    while i < len(top_dict[k]):
        keys.append((top_dict[k][i][0]))
        values.append((top_dict[k][i][1]))
        i+=1
    plt.figure(figsize=(30,15))
    plt.title(k, fontsize=24)
    plt.xlabel('words', fontsize=24)
    plt.ylabel('frequency', fontsize=24)
    plt.bar(keys, values)
print(plt.show())
```

```python
# Look at the most common top words --> add them to the stop word list
from collections import Counter

# Let's first pull out the top 30 words for each comedian
words = []
for rapper in data.columns:
    top = [word for (word, count) in top_dict[rapper]]
    for t in top:
        words.append(t)
        
words
```

```python
# Let's aggregate this list and identify the most common words along with how many routines they occur in
Counter(words).most_common()
```

```python
# If more than half of the rapper songs have it as a top word, exclude it from the list
add_stop_words = [word for word, count in Counter(words).most_common() if count > 0]
add_stop_words
```
