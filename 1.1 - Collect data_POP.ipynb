```python
# function to collect songs in album into array
def collectSongs(song_names, pop):
    lyrics = []
    for song_name in song_names:
        try:
            song = genius.search_song(song_name, pop)
            if song is not None:
                lyrics.append(song)
        except:
            print(song_name)
    return lyrics
```

```python
# raw data dictionary
raw_lyrics_array = {}

# fill raw_data dictionary with song lyrics
for pop in pops:

    if pop == 'Madonna':
        raw_lyrics_array[pop] = collectSongs(madonna_song_names, pop)
    elif pop == 'Michael Jackson':
        raw_lyrics_array[pop] = collectSongs(jackson_song_names, pop)
    elif pop == 'Rihanna':
        raw_lyrics_array[pop] = collectSongs(rihanna_song_names, pop)
    elif pop == 'The Beatles':
        raw_lyrics_array[pop] = collectSongs(beatles_song_names, pop)
    elif pop == 'Taylor Swift':
        raw_lyrics_array[pop] = collectSongs(taylor_song_names, pop)
    else:
        print("No match")
```

```python
# function to collect songs in album into one text
def collectSongsToOneText(song_names, pop):
    text = ""
    for song_name in song_names:
        song = genius.search_song(song_name, pop)
        if song is not None:
            text += song.lyrics + "\n";
    return text
```

```python
# raw data dictionary
raw_data = {}

# fill raw_data dictionary with song lyrics
for pop in pops:

    if pop == 'Madonna':
        raw_data[pop] = collectSongsToOneText(madonna_song_names, pop)
    elif pop == 'Michael Jackson':
        raw_data[pop] = collectSongsToOneText(jackson_song_names, pop)
    elif pop == 'Rihanna':
        raw_data[pop] = collectSongsToOneText(rihanna_song_names, pop)
    elif pop == 'The Beatles':
        raw_data[pop] = collectSongsToOneText(beatles_song_names, pop)
    elif pop == 'Taylor Swift':
        raw_data[pop] = collectSongsToOneText(taylor_song_names, pop)
    else:
      print("No match")
```
